```python
from pathlib import Path
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# datadir = Path('output-s_66-n_50000')
datadir = Path('output-s_42-n_50000')
pimg = "blend_{:06d}.npy"
pseg = "blend_seg_{:06d}.npy"
```

```python
idx = np.random.randint(0, 10000)

img = np.load(datadir / pimg.format(idx))
seg = np.load(datadir / pseg.format(idx))

fig, ax = plt.subplots(ncols=3)
ax[0].imshow(img)
for i in range(2):
    ax[i+1].imshow(seg[i])
```
